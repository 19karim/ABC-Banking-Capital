### Import the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import the Dataset

```python
df = pd.read_csv(r'Train.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.describe(include='all')
```

```python
df.info()
```

### Checking for null values

```python
df.isna().sum()
```

### Checking for unique values of columns

```python
for i in df.columns:
    print({i:df[i].unique()})
```

### Replacing the '?' with nan values

```python
df.replace('?',np.nan,inplace=True)
```

```python
df.isna().sum()
```

### Replacing the 'unknown' with nan values

```python
df.replace('unknown',np.nan,inplace=True)
```

```python
df.isna().sum()/len(df)*100
```

### Filling the nan values with mode and mean

```python
for value in ['account_info','savings_account','duration_month','gurantors','resident_since']:
    df[value].fillna(df[value].mode()[0],inplace=True)
```

```python
for value in ['age']:
    df[value].fillna(df[value].mean(),inplace=True)
```

```python
for value in ['telephone', 'foreigner']:
    df[value].fillna(method='ffill', inplace=True)
```

```python
df.isna().sum()/len(df)*100
```

### Create X and Y

```python
# Features (X): Selecting all columns except 'credit_amount' (target variable)
X = df.drop(columns=['credit_amount'])

# Target variable (y): 'credit_amount'
Y = df['credit_amount']
```

```python
print(X.shape)
print(Y.shape)
```

```python
sns.distplot(Y)
plt.show()
```

```python
colname = []
for i in df.columns:
    if df[i].dtypes=='object':
        colname.append(i)
colname
```

### Converting the catergorical values into numerical values

```python
from sklearn.preprocessing import LabelEncoder
 
le=LabelEncoder()
 
for x in colname:
    X[x]=le.fit_transform(df[x])
```

```python
X.head()
```

### Checlking for multicollinearity using VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
 
vif_df = pd.DataFrame()
vif_df["features"] = X.columns
vif_df["VIF Factor"] = [vif(X.values, i) for i in range(X.shape[1])]
vif_df.round(2)
```

#### Even though the VIF of age is too high will not drop the column because it is significant variable.

```python
X.drop(columns=['telephone','purpose','foreigner'], axis=1, inplace=True)
```

```python
X.shape
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
 
scaler = StandardScaler()
 
scaler.fit(X)
X = scaler.transform(X)
```

### Splitting the train and test data

```python
from sklearn.model_selection import train_test_split
 
#Split the data into test and train
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, 
                                                    random_state=10) 
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

### Model Building

```python
from sklearn.linear_model import LinearRegression
# create a model object
lm = LinearRegression()
# train the model object
lm.fit(X_train,Y_train)
```

```python
# predict using the model
Y_pred = lm.predict(X_test)
print(Y_pred)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
 
r2=r2_score(Y_test,Y_pred)
print("R-squared:",r2)
 
rmse=np.sqrt(mean_squared_error(Y_test,Y_pred))
print("RMSE:",rmse)
 
adjusted_r_squared = 1 - (1-r2)*(len(Y)-1)/(len(Y)-X.shape[1]-1)
print("Adj R-square:",adjusted_r_squared)

normalized_rmse=(1-rmse/100000)*100
print("normalized_rmse",normalized_rmse)
```

```python
from sklearn.linear_model import Ridge

# Create a model object
ridge = Ridge()

# Train the model object
ridge.fit(X_train, Y_train)
```

```python
# predict using the model
Y_pred = ridge.predict(X_test)
print(Y_pred)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
 
r2=r2_score(Y_test,Y_pred)
print("R-squared:",r2)
 
rmse=np.sqrt(mean_squared_error(Y_test,Y_pred))
print("RMSE:",rmse)
 
adjusted_r_squared = 1 - (1-r2)*(len(Y)-1)/(len(Y)-X.shape[1]-1)
print("Adj R-square:",adjusted_r_squared)

normalized_rmse=(1-rmse/100000)*100
print("normalized_rmse",normalized_rmse)
```

```python
from sklearn.linear_model import Lasso

# Create a model object
lasso = Lasso()

# Train the model object
lasso.fit(X_train, Y_train)
```

```python
# predict using the model
Y_pred = lasso.predict(X_test)
print(Y_pred)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
 
r2=r2_score(Y_test,Y_pred)
print("R-squared:",r2)
 
rmse=np.sqrt(mean_squared_error(Y_test,Y_pred))
print("RMSE:",rmse)
 
adjusted_r_squared = 1 - (1-r2)*(len(Y)-1)/(len(Y)-X.shape[1]-1)
print("Adj R-square:",adjusted_r_squared)

normalized_rmse=(1-rmse/100000)*100
print("normalized_rmse",normalized_rmse)
```

### Linear Regression gave the least RMSE so going further with this model.
